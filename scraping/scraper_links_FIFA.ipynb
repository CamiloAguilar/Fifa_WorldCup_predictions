# Scraping web FIFA players
https://www.fifaindex.com/players/

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests
import pandas as pd

url = "https://www.fifaindex.com"
proxies = {
  'http': 'http://10.4.22.5:3128',
  'https': 'https://10.4.22.5:3128',
}

page = requests.get(url)
#page = requests.get(url, proxies=proxies)
```

```python
html = page.content
soup = BeautifulSoup(html,'lxml')
right_table = soup.find('table', class_='table table-striped players')

A=[]
B=[]
C=[]
```

```python
tabla = right_table.findAll("tr")
for row in tabla:
    cells = row.find("td")
    if(cells!=None):
        a = cells.find("a")
        if(a!=None):
            A.append(a["title"])
            B.append(a["href"])
    nations = row.find('img', class_='nation small')
    if(nations!=None):
        C.append(nations["title"])

temp_df=pd.DataFrame({'Name':A, 'Country':C, 'url' : B})
temp_df.head(15)
```

```python
max_page = 603
for i in range(1,max_page):
        if i < 2:
            url_temp = url+'/players/'
            print("Obteniendo página " + str(i))
        else: url_temp = url+'/players/'+str(i)+'/'

        while(True):
            if (i%50 == 0):
                print("Obteniendo página " + str(i))
            try:
                page = requests.get(url_temp)
                #page = requests.get(url_temp,proxies=proxies)
            except requests.exceptions.RequestException as e:  
                print(e)
                continue
            break

        html = page.content
        soup = BeautifulSoup(html,'lxml')
        right_table=soup.find('table', class_='table table-striped players')

        for row in right_table.findAll("tr"):
            cells = row.find("td")
            if(cells!=None):
                a = cells.find("a")
                if(a!=None):
                    A.append(a["title"])
                    B.append(a["href"])
            nations = row.find('img', class_='nation small')
            if(nations!=None):
                C.append(nations["title"])
                    
        df=pd.DataFrame({'Name':A, 'Country':C, 'url' : B})

print("Se han cargado "+ str(max_page) +" páginas")

# Resultado final
df.to_csv('jugadores_enlaces.csv', index = False, encoding = 'utf-8')
df.head(15)
```
